```python
def display_text(texto):
    print(texto)
    input("\nPressione Enter para continuar com a história ...\n")

def display_choices(choices):
    for key, value in choices.items():
        print(f"{key}. {value}")

def get_choice(choices):
    choice = input("Escolha uma das opções acima: ")
    while choice not in choices:
        print("Opção inválida. Tente novamente.")
        choice = input("Escolha uma das opções acima: ")
    return choice
```

```python
def roll_dice():
    return random.randint(1, 20)

def perform_action(step):
    result = roll_dice()
    print(f"Você rolou os dados e tirou o número: {result}")
    if result >= 11:
        display_text(história[step]["success"])
    else:
        display_text(história[step]["failure"])
```

```python
def advance_story(current_step):
    step = história[current_step]
    display_text(step["texto"])

    if "choices" in step and step["choices"]:
        display_choices(step["choices"])
        choice = get_choice(step["choices"])
        next_step = step["next"][choice]
        return next_step
    elif "action" in step:
        perform_action(current_step)
        return "next_chapter" if current_step in ["perdido1", "perdido2"] else "chapter_3"
    else:
        return None
```

```python
def play_game():
    current_step = "início"

    while current_step:
        current_step = advance_story(current_step)

play_game()
```
